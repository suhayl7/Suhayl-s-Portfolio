```python
# Total goals

Total_Goals = players_df['Gls'].sum()
print(Total_Goals)
```

```python
# Calcultaing the total amount of Penalty Goals

penalty_kicks_saved = players_df['PKsv'].sum()
penalty_kicks_missed = players_df['PKm'].sum()
penalty_kicks_attempted = players_df['PKatt'].sum()
Total_PenaltyGoals = penalty_kicks_attempted - (penalty_kicks_saved + penalty_kicks_missed)
print(Total_PenaltyGoals)
```

```python
Total_penalty_attempted = players_df['PKatt'].sum()
print(Total_penalty_attempted)
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13, 6))
penalty_not_scored = penalty_kicks_saved + penalty_kicks_missed
data = [int(penalty_not_scored), int(Total_PenaltyGoals)]
labels = ['penalties Missed', 'Penalties scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct ='%.0f%%')
plt.show
```

```python
# Same as above but fro the Premier League
# Calcultaing the total amount of Penalty Goals

penalty_kicks_saved = prem_players['PKsv'].sum()
penalty_kicks_missed = prem_players['PKm'].sum()
penalty_kicks_attempted = prem_players['PKatt'].sum()
Total_PenaltyGoals = penalty_kicks_attempted - (penalty_kicks_saved + penalty_kicks_missed)
print(Total_PenaltyGoals)

Total_penalty_attempted = prem_players['PKatt'].sum()
print(Total_penalty_attempted)

# Pie chart for penalties missed vs scored
plt.figure(figsize=(13, 6))
penalty_not_scored = penalty_kicks_saved + penalty_kicks_missed
data = [int(penalty_not_scored), int(Total_PenaltyGoals)]
labels = ['penalties Missed', 'Penalties scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct ='%.0f%%')
plt.show

```

```python
# players based on age groups

under21 = players_df[players_df['Age'] < 21]
age21_25 =  players_df[(players_df['Age'] >= 21) & (players_df['Age'] <= 25)]
age26_30 =  players_df[(players_df['Age'] >= 26) & (players_df['Age'] <= 30)]
over30 =  players_df[players_df['Age'] > 30]
```

```python
x = np.array([under21['Player'].count(), age21_25['Player'].count(), age26_30['Player'].count(), over30['Player'].count()])
labels = ["<=21", ">21 & <=25", ">25 & <=30", ">30"]
plt.title('Totla players for each age category' , fontsize = 20)
plt.pie(x, labels = labels, autopct = "%.1f%%")
plt.show()
```

```python
# players based on age groups

under21 = prem_players[prem_players['Age'] < 21]
age21_25 =  prem_players[(prem_players['Age'] >= 21) & (prem_players['Age'] <= 25)]
age26_30 =  prem_players[(prem_players['Age'] >= 26) & (prem_players['Age'] <= 30)]
over30 =  prem_players[prem_players['Age'] > 30]

x = np.array([under21['Player'].count(), age21_25['Player'].count(), age26_30['Player'].count(), over30['Player'].count()])
labels = ["<=21", ">21 & <=25", ">25 & <=30", ">30"]
plt.title('Totla players for each age category' , fontsize = 20)
plt.pie(x, labels = labels, autopct = "%.1f%%")
plt.show()
```

```python
# Average age of players in each team (Premier League)

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Squad', y = 'Age', data = prem_players)
plt.xticks(rotation = 90)
```

```python
num_player = prem_players.groupby('Squad').size()
data = (prem_players.groupby('Squad')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
# Total assist from each club
assists_per_club = pd.DataFrame(prem_players.groupby('Squad', as_index = False)['Ast'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y= 'Ast', data = assists_per_club.sort_values(by='Ast'),palette='Set2')
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Ast", fontsize = 20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title("Total amount of assists per club")
```

```python
# Top 10 Assists 

top_10_assists = prem_players[['Player', 'Squad', 'Ast', '90s']].nlargest(n=10, columns = 'Ast')
top_10_assists

```

```python
# Total Goals from each club
goals_per_club = pd.DataFrame(prem_players.groupby('Squad', as_index = False)['Gls'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y= 'Gls', data = goals_per_club.sort_values(by='Gls'),palette='Set2')
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Gls", fontsize = 20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title("Total amount of Goals per club")
```
